# CO2 Reduction

```python
#The Haversine formula is used to calculate the distance between two points on a sphere given their longitudes and latitudes.
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the earth in km
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat/2) * np.sin(dlat/2) + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon/2) * np.sin(dlon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c
```

```python
txt_files = glob.glob('*.txt')

dataframes = []

for file in txt_files:
    #Load each taxi file into a separate DataFrame with a unique identifier, and convert UNIX epoch time to datetime
    df = pd.read_csv(file, sep=' ', names=['lat', 'lon', 'occupied', 'time'])
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df['taxi_id'] = file
    
    #For each taxi's DataFrame, shift the latitude and longitude columns to create 'previous' coordinate columns for each point
    df['prev_lat'] = df['lat'].shift()
    df['prev_lon'] = df['lon'].shift()
    
    #Apply the haversine function to each taxi's DataFrame row-wise to calculate the distance traveled between consecutive points.
    df['distance_km'] = df.apply(lambda row: haversine(row['prev_lat'], row['prev_lon'], row['lat'], row['lon']), axis=1)
    
    dataframes.append(df)

# Combine all individual taxi DataFrames into a single DataFrame for further analysis.
data = pd.concat(dataframes, ignore_index=True)
```

```python
# Sum all distances travelled by taxis when unoccupied and then normalize it to a per month basis
roaming_distance_per_month_km = data[data['occupied'] == 0]['distance_km'].sum() / (30/len(np.unique(data['time'].dt.date)))

# Convert the roaming distance from kilometers to miles.
roaming_distance_per_month_miles = roaming_distance_per_month_km * 0.621

# Calculate the total CO2 emissions per month based on average emissions per mile.
total_CO2_per_month = 404 * roaming_distance_per_month_miles

# First we calculate the total CO2 emissions for the year considering the initial fleet
total_CO2_per_year = total_CO2_per_month * 12
```

```python
# For each month, calculate the CO2 emissions reduction, considering the proportion of the fleet still using combustion engines
monthly_fleet_change_rate = 0.15
total_reduction = 0

for month in range(1,13):
    # Calculate the proportion of the fleet that's still using combustion engines
    combustion_fleet_proportion = (1 - monthly_fleet_change_rate) ** month
    # Calculate the CO2 emissions for this month, considering the fleet change
    monthly_CO2_combustion = total_CO2_per_month * combustion_fleet_proportion
    # Calculate the reduction in CO2 emissions for this month
    monthly_reduction = total_CO2_per_month - monthly_CO2_combustion
    # Add this month's reduction to the total reduction
    total_reduction += monthly_reduction

# Convert potential CO2 reduction to metric tons for better readibility
total_reduction_tons = total_reduction / 1e6

print("Potential yearly reduction in CO2 emissions caused by taxi cabs roaming without passengers is approximately {:.0f} metric tons.".format(total_reduction_tons))
```

# Taxi Cluster

```python
from sklearn.cluster import KMeans

def find_best_k(data, k_range):
    # Preprocessing
    filtered_data = data.dropna(subset=['lat', 'lon', 'occupied', 'time'])
    coordinates = filtered_data[['lat', 'lon']]

    # Calculate inertia for each k
    inertias = []
    for k in k_range:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(coordinates)
        inertias.append(kmeans.inertia_)

    # Plot the inertia values
    plt.plot(k_range, inertias)
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method: Inertia vs. Number of Clusters')
    plt.show()

    # Find the best k value
    best_k = k_range[0]
    for i in range(1, len(inertias)):
        if inertias[i] - inertias[i-1] < 0.1 * inertias[i-1]:
            best_k = k_range[i-1]
            break

    return best_k

# Define the range of k values to explore
k_range = [4, 5, 6, 7]

# Call the method to find the best k value
best_k = find_best_k(data, k_range)

# Output the best k value
print('Best k:', best_k)
```

```python
# Cluster analysis
coordinates = filtered_data[['lat', 'lon']]
kmeans = KMeans(n_clusters=4)
kmeans.fit(coordinates)
filtered_data['cluster'] = kmeans.labels_

# Analyze the clusters
cluster_summary = filtered_data.groupby('cluster').agg(
    occupied_count=('occupied', 'sum'),
    busiest_time=('time', lambda x: x.dt.hour.value_counts().idxmax()),
    avg_distance=('distance_km', 'mean'),
)
print(cluster_summary)
```
